## STEP 1 引入老師提到的4個基本套件

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## STEP 2 畫圖！
我打算畫一個可愛的狗狗圖，詳細每個部位用到的方式備注在下面程式的註解

```python
# 設定參數
t = np.linspace(0, 2 * np.pi, 300)

# 頭部 (圓形)
head_x = 2 * np.cos(t)
head_y = 2 * np.sin(t) + 3

# 耳朵 (使用貝茲曲線模擬)
ear_left_x = [-1.8, -2.5, -3.0, -2.5, -1.8]
ear_left_y = [3.5, 4.8, 4.5, 3.8, 3.5]
ear_right_x = [1.8, 2.5, 3.0, 2.5, 1.8]
ear_right_y = [3.5, 4.8, 4.5, 3.8, 3.5]

# 眼睛 (小圓)
eye_left_x = 0.3 * np.cos(t) - 0.8
eye_left_y = 0.3 * np.sin(t) + 3.5
eye_right_x = 0.3 * np.cos(t) + 0.8
eye_right_y = 0.3 * np.sin(t) + 3.5

# 鼻子 (圓)
nose_x = 0.25 * np.cos(t)
nose_y = 0.2 * np.sin(t) + 2.4

# 嘴巴 (笑臉拋物線)
mouth_x = np.linspace(-0.4, 0.4, 50)
mouth_y = -0.2 * (mouth_x ** 2) + 2.2

# 舌頭 (橢圓形)
tongue_x = 0.2 * np.cos(t)
tongue_y = 0.4 * np.sin(t) + 1.9
tongue_y = np.clip(tongue_y, 1.9, 2.1)  # 讓舌頭不超出嘴巴

# 畫圖
plt.figure(figsize=(6, 6))

# 使用不同顏色繪製各個部位
plt.fill(head_x, head_y, color="#F4A460", label="Head")  # 頭部
plt.fill(ear_left_x, ear_left_y, color="brown", label="Left Ear")  # 左耳
plt.fill(ear_right_x, ear_right_y, color="brown", label="Right Ear")  # 右耳
plt.fill(eye_left_x, eye_left_y, color="black", label="Left Eye")  # 左眼
plt.fill(eye_right_x, eye_right_y, color="black", label="Right Eye")  # 右眼
plt.fill(nose_x, nose_y, color="black", label="Nose")  # 鼻子
plt.plot(mouth_x, mouth_y, color="black", linewidth=2)  # 嘴巴
plt.fill(tongue_x, tongue_y, color="red", label="Tongue")  # 舌頭

# 調整圖形
plt.xlim(-3, 3)
plt.ylim(-1, 5)
plt.axis("on")
plt.grid()
plt.title("Cute dog", fontsize=14)

# 顯示圖形
plt.show()
```
